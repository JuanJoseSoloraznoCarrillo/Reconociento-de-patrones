### 3.3.5 **Dictionaries.**

Es una estructura que permite almacenar varios tipos de datos oredenados en pares (clave-valor)




```
#Diccionario vacío
empty_dict = {} 
empty_dict2 = dict ()

# Dos formas de crear un diccionario
family = {'dad':'homer', 'mom':'marge', 'size':6}
family = dict(dad='homer', mom='marge', size=6)

#Convertir una lista de tuplas a un diccionario
list_of_tuples = [('dad','homer'), ('mom','marge'), ('size', 6)]
family = dict(list_of_tuples)

#Examinar un diccionario
family['dad'] # regresa el elemento:  'homer'
len(family) # regresa el tamaño del diccionario:  3
family.keys() # regresa un lista con las llaves del dicionario: ['dad', 'mom', 'size']
family.values() # regresa una lista con los valores de las llaves de un diccionario: ['homer', 'marge', 6]
fam = family.items() # regresa un arreglo de tuplas
print(fam)
'mom' in family # revisa el nombre de las llaves y si encuentra mom retorna un verdadero.
'marge' in family # si no regresa un falso

#Modificar un diccionario
family['cat'] = 'snowball' #Agrega un nuevo par de valores al diccionario.
family['cat'] = 'snowball ii' # renombramos el valor de la llave "cat"
del family['cat'] # Eliminamos elementos de un diccionario.
family['kids'] = ['bart', 'lisa'] # Agregamos valores en forma de lista
family.update({'baby':'maggie', 'grandpa':'abe'}) # Agregamos multiples pares al diccionario.
family['kids'][0] # regresa de la llave "kids" el primer elemento: 'bart'
family['kids'].remove('lisa') # elimina de la llave kids el elemento 'lisa'


```

### 3.3.5 **Set.**

Es como un diccionario pero solo con claves únicas, pueden contener varios tipos de valores como cadenas números o tuplas.

```
# crear un set vacío
empty_set = set()
# crear un set
languages = {'python', 'r', 'java'}
snakes = set(['cobra', 'viper', 'python']) # crear un set desde una lista

# revisar un set
len(languages) # regresa el tamaño del set: 3
'python' in languages # busca el elemento 'python' y si existe en el set, regresa un verdadero.

# operaciones con sets
languages & snakes # regresa la intersección de los sets: {'python'}
languages | snakes # regresa la unión de los sets: {'cobra', 'r', 'java', 'viper', 'python'}
languages - snakes # devuelve la diferencia: {'r', 'java'}
snakes - languages # devuelve la diferencia: {'cobra', 'viper'}

# modicar un set
languages.add('sql') # agrega un nuevo elemento al set
languages.add('r') 
languages.remove('java') # elimina un elemento del set
```

### 3.4.2 **Loops.**


```
# range retorna una lista de enteros

# retorna [0, 1, 2]
range(0, 3)
range(3) 

# Ciclo for
fruits = ['apple', 'banana', 'cherry']

for i in range(len(fruits)):
  print(fruits[i].upper()) #Imprime el arreglo fruits en mayusculas

family = {'dad':'homer', 'mom':'marge', 'size':6}

#ciclo utilizando tuplas
for key, value in family.items():
  print(key, value)

  # ciclo for, if-else
for fruit in fruits: 
  if fruit == 'banana':
    print("Found the banana!")
    break # sale del ciclo y salta al else
  else:
      print("Can't find the banana")

# Ciclo while
count = 0
while count < 5:
  print("This will print 5 times")
  count += 1 #Contador del ciclo
```

###3.4.3 **Iteradores y lista de compresiones.**

Este proceso afecta a listas completas sin iterar a través de bucles. 

```
# crear una lista de números
nums = [1, 2, 3, 4, 5]
cubes = [] #lista vacía para los cubos del los números de la lista nums

for num in nums:
  cubes.append(num**3)

# Crear directamente la lista de cubos de los números sin el ciclo for
cubes2 = [num**3 for num in nums] # [1, 8, 27, 64, 125]

print(cubes)
print(cubes2)

#convertir matriz 2d en vector
matrix = [[1, 2], [3, 4]]
items = []
for row in matrix:
  for item in row:
    items.append(item)

#equivalencia de utilizar una lista de compresión    
items2 = [item for row in matrix
for item in row]

print(items)
print(items2)
```
