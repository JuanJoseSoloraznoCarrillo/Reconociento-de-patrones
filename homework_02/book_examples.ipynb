```
# CREAR DATAFRAME
from __future__ import print_function
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Crear tabla de contenido
columns = ['Nombre', 'Edad', 'Genero', 'Ocupacion']  #Encabezado de  la tabla 

#Contenido de la tabla
usuario_1 = pd.DataFrame([['alice', 19, "F", "student"],['john', 26, "M", "student"]],columns=columns)
usuario_2 = pd.DataFrame([['eric', 22, "M", "student"],['paul', 58, "F", "manager"]],columns=columns)
usuario_3 = pd.DataFrame(dict(Nombre=['peter', 'julie'],Edad=[33, 44], Genero=['M', 'F'],Ocupacion=['engineer', 'scientist']))

print(usuario_1)

users = pd.concat([usuario_1, usuario_2, usuario_3])
print(users)


#DataFrame de formato largo a formato ancho
staked = pd.melt(users, id_vars="Nombre", var_name="variable", value_name="valor")
print(staked)



```

```
#SORTING

df = users.copy()

#df.age.sort_values() # only works for a Series

df.sort_values(by='Edad') # Ordena las filas por columnas específicas
df.sort_values(by='Edad', ascending=False) # ordenación desendente
df.sort_values(by=['Ocupacion', 'Edad']) # ordenacion de varias columnas
df.sort_values(by=['Ocupacion', 'Edad'], inplace=True) # modifica df
print(df)
```

```
#DATOS DUPLICADOS

df = users.append(df.iloc[0], ignore_index=True)

print(df.duplicated()) # imprime los valores booleanos

df.duplicated().sum() # cuenta los duplicados

df[df.duplicated()] # muestra solo los que se duplica

df = df.drop_duplicates() # filas duplicadas
```
